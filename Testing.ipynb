```python

x = np.array([0.5,0.2, 0.7])
y = np.array([0.2,0.2])
r = np.array([0.5, 0.7, 0.3])

phi = np.linspace(0.0,2*np.pi,100)

na=np.newaxis

# the first axis of these arrays varies the angle, 
# the second varies the circles
x_line = x[na,:] + r[na,:]*np.sin(phi[:,na])
y_line = r[na,:]*np.cos(phi[:,na])

plt.plot(x_line,y_line,'-')
plt.grid()
plt.show()
```

```python
"""Constants to be used in subsequent code. """
o_gamma = 6.67384 * (10**-11) #m3 kg-1 s-2;    tu=10^8 years
gamma = o_gamma* (3.24077929e-20/1)**3 * (1/3.1689e-8)**2 * (1.989e30/1) * (10**8/1)**2 #kpc3 m_sun-1 tu-2
gamma
#         Vold = np.sqrt((2.0 * gamma * M/r_s[i]) + ((2*(e**2 - 1)*gamma**2 * M**2 * S**2)/(2*L**2)))
```

```python
def initial_galaxy(M, S, e):
    icg = []
    Y = np.linspace(-50, 50, 120)
    if e == 1:
        X = []
        for j in range(len(Y)):
            X.append(Y[j]**2/100 - 25)
        for i in range(len(X)):
            r_s = np.sqrt(np.array(X)**2 + np.array(Y)**2)

            phi = np.arctan(4.0*25/(-2.0*np.array(Y)))
    #         alpha = np.arctan(np.array(Y)/np.array(X))

    #         a = np.array(r_s) * (1 + e*np.cos(np.array(alpha))) / (1 - e**2)
            V = np.sqrt((2*gamma*(S+M))/np.array(r_s))# - (gamma*M)/np.array(a))

            Vx = np.array(V)*np.cos(np.array(phi))
            Vy = np.array(V)*np.sin(np.array(phi))

            IC = np.array([X[i], Y[i], Vx[i], Vy[i]])
            icg.append(IC)
    else:
        return False
    return icg
```

```python
M, S, e = 1e11, 1e11, 1
ic_gala = initial_galaxy(M, S, e)
# ic_gala
```

```python
x = [p[0] for p in ic_gala]
y = [p[1] for p in ic_gala]
vx = [p[2] for p in ic_gala]
vy = [p[3] for p in ic_gala]
```
